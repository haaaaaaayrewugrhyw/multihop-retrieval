```python
# Cell A - PARAPHRASE aux (surface-invariance)
!pip install transformers scikit-learn scipy datasets -q
import subprocess, sys, os
from pathlib import Path

# Load HF token in the MAIN kernel (kaggle_secrets is unreliable inside a subprocess).
# The subprocess inherits os.environ, so the token reaches the script -> no rate-limit stall.
try:
    from kaggle_secrets import UserSecretsClient
    _t = UserSecretsClient().get_secret("HF_TOKEN")
    os.environ["HF_TOKEN"] = _t; os.environ["HUGGING_FACE_HUB_TOKEN"] = _t
    print("HF token: loaded in kernel")
except Exception as e:
    print("HF token: NONE (add a Kaggle secret named HF_TOKEN) ->", e)

REPO = Path('/kaggle/working/multihop-retrieval')
if not REPO.exists():
    os.system(f'git clone https://github.com/haaaaaaayrewugrhyw/multihop-retrieval.git {REPO}')
else:
    print(os.popen(f'git -C {REPO} pull').read())
subprocess.run([sys.executable, '-u', str(REPO/'delta_system'/'delta2_model.py'),
                '--aux', 'paraphrase', '--steps', '400'])   # inherits HF_TOKEN from os.environ
```

```python
# Cell B - NLI aux (delta carries the entail/neutral/contradict relation)
import subprocess, sys, os
from pathlib import Path
try:
    from kaggle_secrets import UserSecretsClient
    _t = UserSecretsClient().get_secret("HF_TOKEN")
    os.environ["HF_TOKEN"] = _t; os.environ["HUGGING_FACE_HUB_TOKEN"] = _t
    print("HF token: loaded in kernel")
except Exception as e:
    print("HF token: NONE ->", e)
REPO = Path('/kaggle/working/multihop-retrieval')
subprocess.run([sys.executable, '-u', str(REPO/'delta_system'/'delta2_model.py'),
                '--aux', 'nli', '--steps', '400'])
```
