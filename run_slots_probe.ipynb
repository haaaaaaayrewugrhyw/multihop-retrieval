# Hard slot bottleneck + scattered data (last PURE shot)

Combines the two lessons: scattered training makes delta NEEDED for novel (no left-context
shortcut), and a hard K-slot bottleneck makes delta UNABLE to also carry shared (it must
spend its few slots on what A lacks = novelty). Pure — no labels, no oracle.

Success = the **specialization gap** (novel vs shared full-effect; baseline = -0.14) flips
toward/positive, and ideally **A-dependence** rises off ~6-8%.

Saves to `wiki_model_wae_slots.pt`; baselines untouched. GPU ON, Internet ON.
